```python
import random
import matplotlib.pyplot as plt

n_values = 100
y_values = []


for i in range(n_values):
    y_values.append(random.randint(1, 100))

x_values = range(n_values) 
plt.plot(x_values, y_values)
plt.show()
```

```python
n_values = 100
y_values_1 = []
y_values_2 = []

# Create data (y_values) randomly between 1 and 100.
for i in range(n_values):
    y_values_1.append(random.randint(1, 100))
    y_values_2.append(random.randint(1, 100))

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values_1)
plt.plot(x_values, y_values_2)  # call plot again draws in the same graph.
plt.show()
```

```python
x_values = list(range(10))              # [0, 1, 2, ..., 9]
y_values = [i * 2 for i in x_values]    # [0, 2, 4, ..., 18]

plt.plot(x_values, y_values, marker='o')
plt.title('Ascending Line')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
x_values = list(range(10))
y_values = [20 - i * 2 for i in x_values]  # [20, 18, 16, ..., 2]

plt.plot(x_values, y_values, marker='o', color='red')
plt.title('Descending Line')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
import math
x_values = [i for i in range(100)]                   # 100 values
y_values = [math.sin(i * 0.1) for i in x_values]     # Sine wave

plt.plot(x_values, y_values)
plt.title('Wave Pattern using Sine')
plt.xlabel('X Axis')
plt.ylabel('Y Axis (sin)')
plt.grid(True)
plt.show()
```
