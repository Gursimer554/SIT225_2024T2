```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
acc_x = pd.read_csv("C:\\Users\\gursi\\Downloads\\historic-data-20250915T031615Z\\OnePlus 10R 5G Thing-Accelerometer_X.csv")
acc_y = pd.read_csv("C:\\Users\\gursi\\Downloads\\historic-data-20250915T031615Z\\OnePlus 10R 5G Thing-Accelerometer_Y.csv")
acc_z = pd.read_csv("C:\\Users\\gursi\\Downloads\\historic-data-20250915T031615Z\\OnePlus 10R 5G Thing-Accelerometer_Z.csv")


x_data = acc_x.iloc[:, 1]  # 2nd column
y_data = acc_y.iloc[:, 1]
z_data = acc_z.iloc[:, 1]


plt.figure(figsize=(12, 6))
plt.plot(x_data, label="Accelerometer X", color="blue")
plt.title("Accelerometer X - Line Graph (Circular/Oval Motion)")
plt.xlabel("Samples")
plt.ylabel("X-axis Acceleration")
plt.legend()
plt.show()

# Plot accelerometer_y (Bar plot)
plt.figure(figsize=(12, 6))
plt.bar(range(len(y_data)), y_data, color="green")
plt.title("Accelerometer Y - Bar Plot (Circular/Oval Motion)")
plt.xlabel("Samples")
plt.ylabel("Y-axis Acceleration")
plt.show()

# Plot accelerometer_z (Histogram)
plt.figure(figsize=(12, 6))
plt.hist(z_data, bins=30, color="orange", alpha=0.7)
plt.title("Accelerometer Z - Histogram (Circular/Oval Motion)")
plt.xlabel("Z-axis Acceleration")
plt.ylabel("Frequency")
plt.show()

# Combined plot with all 3 (Line Graph)
plt.figure(figsize=(12, 6))
plt.plot(x_data, label="Accelerometer X", color="blue")
plt.plot(y_data, label="Accelerometer Y", color="green")
plt.plot(z_data, label="Accelerometer Z", color="red")
plt.title("Combined Accelerometer X, Y, Z - Line Graph (Circular/Oval Motion)")
plt.xlabel("Samples")
plt.ylabel("Acceleration")
plt.legend()
plt.show()
```
